```python
import matplotlib.pyplot as plt #繪圖函式庫
import numpy as np #矩陣函式庫
from sklearn import datasets,linear_model #線性邏輯函式庫

#取得糖尿病的數據
diabetes = datasets.load_diabetes()

#只取第三個特徵值BMI
diabetes_X = diabetes.data[:,np.newaxis,2]

#切分三個特徵值BMI
diabetes_X_train = diabetes_X[:-20] #切割0到最後20筆特徵給訓練用
diabetes_X_test = diabetes_X[-20:] #切割最後20筆特徵給訓練用

#切分答案
diabetes_Y_train = diabetes.target[:-20] #切割0到最後20筆特徵給訓練用
diabetes_Y_test = diabetes.target[-20:] #切割最後20筆特徵給訓練用

#計算
regr = linear_model.LinearRegression() #建立線性邏輯

regr.fit(diabetes_X_train,diabetes_Y_train) #訓練

#係數
print('Coefficient:\n',regr.coef_)
#均方誤差
print("Mean squared error:%.2f"
      %np.mean((regr.predict(diabetes_X_test)-diabetes_Y_test)**2))

#顯示方差分數:1是完美預測
print('Variance score:%.2f'%regr.score(diabetes_X_test,diabetes_Y_test))

#繪圖
plt.scatter(diabetes_X_test,diabetes_Y_test,color='black') #畫出測試黑點
plt.plot(diabetes_X_test,regr.predict(diabetes_X_test),color='blue',linewidth=3) #劃出測試預測的線性回歸答案

plt.xticks(())
plt.yticks(())

plt.show() #顯示
```
